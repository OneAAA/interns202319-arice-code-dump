```python
## Set variance
variance = '0.01'

## Parses diffusion stuff
LiLi_bin = open('Different-Variances/v='+variance+'/Diff_RDFs/LiLiBins', 'r')
LiLi_res = open('Different-Variances/v='+variance+'/Diff_RDFs/LiLiResults', 'r')

def parse(in_stream, delim):
    out_arr = []
    in_arr = in_stream.split(delim)
    for str_num in in_arr:
        if len(str_num) == 0:
            continue
        out_arr.append(float(str_num))
    return out_arr

LiLi_x = parse(LiLi_bin.read(), '\n')
LiLi_y = parse(LiLi_res.read(), '\n')

LiLi_bin.close()
LiLi_res.close()

import matplotlib.pyplot as plt

plt.plot(LiLi_x, LiLi_y)
plt.xlim([0,12])
plt.ylim([0,1.5])
plt.show()
```
